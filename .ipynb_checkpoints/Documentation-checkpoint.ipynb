 # Indoor Localization and Navigation


# Background
Indoor localization and navigation systems are in great demand due to its potential use cases in the business. We can easily localize and navigate in the outdoor environment because of the availability of accurate GPS system but the accuracy of this system is highly dependent on the visibility of GPS satellites. In the indoor environment, it’s accuracy degrades to 50 meters which makes it not useful for indoor applications. To localize in the indoor environment we need to use some other techniques along with additional sensors like inertial measurement unit (IMU), radio frequency identification (RFID), WiFi and Bluetooth low energy (BLE).  
From the past few decades, Inertial Navigation System (INS) technology is being used for indoor navigation. Previously, quite large and heavy gyroscope and accelerometer were used to sense the rotations and accelerations which were converted into position and orientation using the algorithm known as strapdown navigator. But now the availability of the micro-electromechanical system (MEMS) based IMU has reduced the form factor significantly. Now we can get the IMU of a few grams and a few centimeters in size. The problem with the MEMS-based IMU is that they are inherently inaccurate.  
The gyroscope gives Agular velocity and accelerometer gives acceleration, in the strapdown algorithm, we integrate the gyroscope readings and double integrate the accelerometer readings to find orientation and position respectively. Because of inherent error in MEMS-based IMU and integration results in the error accumulation which grows exponentially with time.

# Inertial Navigation System
An INS is a self-contained Dead Reckoning (DR) navigation system, which provides
dynamic information about position and velocity, based on the measurements obtained
from an Inertial Measurement Unit (IMU). We are using MPU6881 which is 6DOF IMU which means it contains three accelerometers and three gyroscopes mounted on a triad.  The basic operating principle of inertial navigation is based on Newton’s law of motion, which says that an
object continues to be in a state of rest or uniform motion unless acted upon by an
external force. The application of any external force generates the acceleration, which is
sensed by accelerometers contained in an IMU. This acceleration, when integrated twice,
provides the change in the state of the object with respect to the initial conditions. Thus,
an INS generically provides the change in the state of the object, and the knowledge of
the initial condition (position, velocity, and attitude) of the platform prior to the start of
the navigation is essential for its operation.   
However, the computation process is more complicated then it sounds because of the fact
that the IMU provides measurements in a coordinate frame different than the coordinate
frame in which position and velocity are usually expressed. This is what
drives the need of gyros in an INS, which provides angular measurements to be able to
transform the sensed acceleration to an appropriate frame, prior to integration. It follows
that the accuracy of the angular measurements is fundamental to an INS because any
errors in the transformation of acceleration will ultimately lead to errors in position. Thus,
the ability of an INS to enable the continuous determination of device position, velocity
and attitude, primarily depends on the quality of gyro sensors used.  

## 1. Coordinate Frames
A set of three noncoplanar and mutually perpendicular axes is used to describe the orientation, position and motion of an object. This position, orientation, and motion of the object are calculated with respect to a reference frame which is also defined by its origin and a set of axes. Suitable origin can be the center of the earth, the center of the solar system, and convenient
local landmarks. Suitable axes include the north, east, and vertical directions; the
Earth’s axis of rotation and vectors within the equatorial plane; the alignment of a
local road grid; the walls of a building; and a line joining two landmarks. Another
object may also act as a reference.  
The origin and axes of either an object or a reference collectively comprise a
coordinate frame. When the axes are mutually perpendicular, the coordinate frame
is orthogonal and has six degrees of freedom. A coordinate frame may be used to describe either an object or a reference.   
Any navigation problem involves at least two coordinate frames. These are
the object frame, describing the body whose position and/or orientation is desired,
and the reference frame, describing a known body, such as the Earth, relative to
which the object position and/or orientation is desired.  
Our INS requires transformations between four coordinate systems, which are defined below. 

### a) Earth-Centered Inertial Frame (ECI or i-frame)
It is a non-rotating and non-accelerating frame with respect to fixed stars. The inertial frame is defined as follows:  

* Origin: Earth’s center of mass
* Zi-Axis: Parallel to the spin axis of the Earth
* Xi-Axis: Pointing towards the mean vernal equinox
* Yi-Axis: Orthogonal to X and Z completing a right-handed system  

### b) Earth Centered Earth Fixed Frame (ECEF or e-frame)
It is defined as follows.  
* Origin: Earth’s center of mass
* Ze-Axis: Parallel to the mean spin axis of the Earth
* Xe-Axis: Pointing towards the mean meridian of Greenwich
* Ye-Axis: Orthogonal to the X and Z axes to complete a right-handed frame

### c) Local Level Frame (LLF or l-frame)
It is also called the navigation frame (n-frame) or geodetic frame (lat, lon, alt). It is defined as follows.
* Origin: Coinciding with sensor frame
* Zl-Axis: Orthogonal to reference ellipsoid pointing Up
* Xl-Axis: Pointing towards geodetic East41
* Yl-Axis: Pointing toward geodetic North

![alt text](./images/llfvsecef.png)

### d) Body Frame (b-frame)
Body frame is an orthogonal frame, whose axes coincide with the axis of the IMU. The body frame is assumed to be aligned with the device frame, and is defined as follows:
* Origin: Centre of the IMU
* Xb-Axis: Pointing towards the right of the vehicle
* Yb-Axis: Pointing towards the front of the vehicle
* Zb-Axis: Orthogonal to the X and Y axes to complete a right-handed system

## 2. Orientation Representations


### a) Coordinate Transformation Matrix
The coordinate transformation matrix, or rotation matrix, is a 3x3 matrix, denoted as \\(C^{\beta }_{\alpha }\\). This is an orthogonal matrix and which belongs to the *Special Orthogonal Group or order 3 (SO(3))* by fulfilling all of the properties of this matrix group.  
A vector may be transformed in one step from one set of resolving axes to another by pre multiplying it by the appropriate coordinate transformation matrix. Thus, for an arbitrary vector, $\pmb x$,
\\[\pmb x^{\beta}=C^{\beta }_{\alpha }\pmb x^{\alpha}\\]

The superscript of $\pmb x$ denotes the resolving axes (the observer’s frame). The lower index \\(\alpha\\) of the matrix \\(C^{\beta }_{\alpha }\\) represents the “from” coordinate frame and the upper index \\(\beta\\) shows the “to” frame.   
The \\(C^{\beta }_{\alpha }\\) matrix can be made using the Euler Angle Representations which states that the rotational motion of a rigid body is completely defined by tracking the body frame with referene to a fixed frame using a set of three Euler angles roll ($\phi$) pitch ($\theta$), and yaw($\psi$). These three angles are used to make corresponding rotation matrices \\(R_{x}\\), \\(R_{y}\\) and \\(R_{z}\\) representing the rotations along x, y and z-axis respectively. The transformation matrix is given below.
$$C^{\beta }_{\alpha } = R_xR_yR_z$$

$$C^{\beta}_{\alpha} = \left[\begin{array}{cc} 
1 & 0 & 0\\
0 & cos\phi & sin\phi\\
0 & -sin\phi & cos\phi
\end{array}\right]
\left[\begin{array}{cc} 
cos\theta & 0 & -sin\theta\\
0 & 1 & 0\\
sin\theta & 0 & cos\theta
\end{array}\right]
\left[\begin{array}{cc} 
cos\psi & sin\psi & 0\\
-sin\psi & cos\psi & 0\\
0 & 0 & 1
\end{array}\right]$$

The above matrix can be used to transform the resolving axes from reference frame to body frame but we need the inverse of it. The transformation of the resolving axes from body frame to inertial frame can be obtained by taking the inverse of \\(C^{\beta}_{\alpha}\\). As this transformation matrix belongs to *SO(3)* so the inverse \\(C^{\alpha }_{\beta}\\) can be obtained by taking the transpose of \\(C^{\beta}_{\alpha}\\) matrix.  
We can also get three Euler angles from \\(C^{\alpha }_{\beta}\\) using the below relations.
$$\phi = arctan(C^{\alpha }_{\beta2,3}, C^{\alpha }_{\beta3,3})$$
$$\theta = -arcsin(C^{\alpha }_{\beta 1,3})$$
$$\psi = arctan(C^{\alpha }_{\beta1,2}, C^{\alpha }_{\beta1,1})$$

## 3. Initial Alignment
One of the important issues while working with a low-cost MEMS IMU is the initial alignment of the system. The process of computing the initial parameters (roll, pitch, yaw) of the transformation matrix is called the INS alignment procedure. The roll and pitch angles can be estimated from accelerometer and yaw from the magnetometer. 
### a) Roll and Pitch Estimation from 3-axis Accelerometer
The roll and pitch of the device can be obtained using the stationary accelerometer when no forces other than gravity act on the device and in the absence of sensor noise. Unfortunately, this is not really the case in practice so the estimated angle will also be quite noisy. However, the gravity vector provides a stable reference direction that remains constant over
time so even if individual measurements and angle estimates are noisy, there is no drift.  

Let's suppose we have accelerometer data $\pmb a = (a_x, a_y, a_z)$ then the roll and pitch can be estimated using Euler angles roll ($\phi$) pitch ($\theta$), and yaw($\psi$) that represents the rotations around $x, y$ and $z$ axis respectively. We are using fixed axis rotations with the order of rotaion as yaw, pitch and then roll. Therefore,


$$\left[\begin{array}{cc} 
a_x\\
a_y\\
a_z
\end{array}\right] = \left[\begin{array}{cc} 
1 & 0 & 0\\
0 & cos\phi & sin\phi\\
0 & -sin\phi & cos\phi
\end{array}\right]
\left[\begin{array}{cc} 
cos\theta & 0 & -sin\theta\\
0 & 1 & 0\\
sin\theta & 0 & cos\theta
\end{array}\right]
\left[\begin{array}{cc} 
cos\psi & sin\psi & 0\\
-sin\psi & cos\psi & 0\\
0 & 0 & 1
\end{array}\right]
\left[\begin{array}{cc} 
0\\
0\\
g
\end{array}\right]$$

Here, the most right vector is the gravitational vector in the inertial frame of reference being transformed in the boday frame of IMU. As accelerometer is insensitive to yaw rotations so it is useless to include $R_z$. The final result after the multiplication is given below


$$\left[\begin{array}{cc} 
a_x\\
a_y\\
a_z
\end{array}\right]= \left[\begin{array}{cc} 
 -gsin\theta \\
gsin\phi \space cos\theta\\
gcos\phi \space cos\theta
\end{array}\right]$$

Roll and pitch angles can be calculated from the above set of equations. On the other hand, the heading angle (yaw) can be estimated from magnetometer using the North vector as a reference. After calculating these three initial Euler angles, we can use them in the rotation matrices of $R_x, R_y, R_z$ and initialize our ${C^{i }_{b}}$.  
For the time being, we don’t have a magnetometer in our IMU so we are initializing all the values of the roll, pitch and yaw with zero degrees and resultant initial \\(C^{i }_{b}\\) becomes an identity matrix.



## 4. Inertial Frame Navigation Equations
We can implement the navigation solution in three different coordinate frames like in ECI, ECEF, and LLF. Every implementation has its own pros and cons. We have implemented our solution in the ECI frame because of its simplicity in implementation and theoratical understanding. ECI is simple because it remains fixed and our IMU gives values in its body frame with reference to the ECI frame. Moreover, by considering our application, we have aignored the earth's rotation in the implementation. In this case, \\(C^{\alpha }_{\beta}\\) is our transformation matrix to convert IMU data from body frame of the IMU to ECI frame, where \\(\alpha\\) is the i-frame and \\(\beta\\) is the b-frame hence it becomes \\(C^{i }_{b}\\).
There are four steps involved in the inertial frame implementation which are described below.

### a) Attitude Update
The attitude update step of the inertial navigation equations uses the angular-rate measurement from the IMU,  \\(\omega^{b}_{ib}\\), to update the attitude solution, expressed as the body-to-inertial-frame coordinate transformation matrix, \\(C^{i }_{b}\\).
The time derivative of the coordinate transformation matrix is  

\\[\dot{C^{i }_{b}} = C^{i }_{b}\Omega^{b }_{ib}\\]
Where $\Omega^{b }_{ib}$ is the angular velocity of body frame b with reference to inertial frame i and observed in body frame b.  
The attitude update equation with the assumption of constant anguar rate over the attitude integration interval is given as
\\[C^{i }_{b}(+) = C^{i }_{b}(-)(I_{3} + \Omega^{b}_{ib} \tau_{i})\\]
Where
$$I_{3} + \Omega^{b}_{ib} \tau_{i} = \left[\begin{array}{cc} 
1 & -\omega^{b}_{ib,z}\tau_{i} & \omega^{b}_{ib,y}\tau_{i}\\
\omega^{b}_{ib,z}\tau_{i} & 1 & -\omega^{b}_{ib,x}\tau_{i}\\
-\omega^{b}_{ib,y}\tau_{i} & \omega^{b}_{ib,x}\tau_{i} & 1
\end{array}\right]$$

### b) Specific-Force Frame Transformation  
The IMU measures specific force along the body-frame resolving axes. However, for
use in the velocity integration step of the navigation equations, it must be resolved
about the same axes as the velocity—in this case, an ECI frame. The resolving axes
are transformed simply by applying a coordinate transformation matrix:
$$ f^{i}_{ib} = C^{i}_{b}(t)f^{b}_{ib} $$

As the specific-force measurement is an average over time t to t + ti, the coordinate transformation matrix should be similarly averaged. A simple implementation,
assuming a constant angular rate is  
$$ f^{i}_{ib} = \frac{1}{2}(C^{i}_{b}(-) + C^{i}_{b}(+)f^{b}_{ib}) $$
However, the mean of two coordinate transformation matrices does not precisely
produce the mean of the two attitudes.  

### c) Velocity Update  
Inertially referenced acceleration is obtained simply by adding the gravitational acceleration to the specific force:  
$$ a^{i}_{ib} = f^{i}_{ib} + \gamma^{b}_{ib}(r^{i}_{ib}) $$
where models the gravitational acceleration, $\gamma^{b}_{ib}$, as a function of Cartesian
position in an ECI frame. 
When the reference frame and resolving axes are the same, the time derivative
of velocity is simply acceleration. Thus,  
$$ \dot{v}^{i}_{ib} = a^{i}_{ib}$$
When variations in the acceleration over the velocity update interval are not
known, as is the case when the velocity integration is iterated at the IMU output
rate, the velocity update equation, obtained by integrating  
$$ v^{i}_{ib}(+) = v^{i}_{ib}(-) + a^{i}_{ib}\tau_{i} $$

### d) Position Update  
In the inertial-frame implementation of the navigation equations, the time derivative of the Cartesian position is simply velocity as the reference frame and resolving
axes are the same. Thus,  
$$ \dot{r}^{i}_{ib} = v^{i}_{ib}$$
In the velocity update step where the variation in acceleration is unknown, $v^{i}_{ib}$
is typically modeled as a linear function of time over the interval $t$ to $t + \tau_{i}$. Integrating above thus leads to the position being modeled as a quadratic function of time.
The velocity is known at the start and finish of the update interval, so the position
is updated using  

$$ r^{i}_{ib}(+) = r^{i}_{ib}(-) + v^{i}_{ib}(-)\tau_{i} + a^{i}_{ib}\frac{\tau^{2}_{i}}{2} $$

# 4 Sensor Errors

As discussed earlier, a gyroscope measures the sensor’s angular velocity ${\omega^b}_{ib}$ ($\frac{deg}{s}$), i.e. the rate of change of the sensor’s orientation. An accelerometer measures the external specific force $f^b$ ($\frac{m}{s^2}$) acting on the sensor. The specific force consists of both the sensor’s acceleration and the earth’s gravity. 

Integration of the gyroscope measurements provides information about the orientation of the sensor. After subtraction of the earth’s gravity, double integration of the accelerometer measurements provides information about the sensor’s position. To be able to subtract the earth’s gravity, the orientation of the sensor needs to be known. Hence, estimation of the sensor’s position and orientation are inherently linked when it comes to inertial sensors. The process of integrating the measurements from inertial sensors to obtain position and orientation information, often called dead-reckoning as shown in the figure below.
![Schematic of gyroscope and accelerometer integration](./images/Integration_gyro_accel.png)

If the initial pose would be known, and if perfect models for the inertial sensor measurements would exist, the process illustrated in Figure above would lead to perfect pose estimates. In practice, however, the inertial measurements are noisy and biased. Because of this, the integration steps from angular velocity to rotation and from acceleration to position introduce integration drift.

Two of these reasons are a slowly time-varying sensor bias and the presence of measurement noise. Let us now assume that we measure we use a non-perfect sensor. In this case our measurements are corrupted by a constant bias, integration of these measurements will lead to a signal which grows linearly with time. Double integration leads to a signal that instead grows quadratically with time. If the sensor instead measures a zero-mean white noise signal, the expected value of the integrated measurements would be zero, but the variance would grow with time. The following example shows the integration drift by assuming a signal $y_t = e_t$, where $e_t = \mathcal{N(0,1)}$ and integrating it for 50 realizations.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
yt_50 = np.random.normal(0,1,50)
y50_ls = [yt_50]
type(y50_ls)
```

```python
# Draw 50 samples from a normal distribution
yt_50 = np.random.normal(0,1,50)
y50_ls = [yt_50]

# integrate the 50 realizations
for i in range(len(yt_50)):
    y50_ls.append(y50_ls[i] + np.random.normal(0,1,50))
    
y50_ls = np.array(y50_ls)
print(y50_ls)
```

```python
# plot all 50 realizations
y50_ls = y50_ls.T
for i in range(50):
    plt.plot(y50_ls[i])
plt.show
```

## 4.1 Angular Velocity

The gyroscope measures the angular velocity of the body frame with respect to the inertial frame,
expressed in the body frame, denoted by ${\omega^b}_{ib}$

\\[ {\omega^b}_{ib} = R^{bn}(\omega^n_{ie} + \omega^n_{en}) + \omega^b_{nb} \\]

where $R_{bn}$ is the rotation matrix from the navigation frame to the body frame. The earth rate, i.e.
the angular velocity of the earth frame with respect to the inertial frame is denoted by $\omega_{ie}$ . The earth rotates around its own z-axis in 23.9345 hours with respect to the stars. Hence, the earth rate is
approximately $7.29 · 10^{-5}$ rad/s. If the navigation frame is not defined stationary with respect to earth than the transport rate i.e. $\omega_{en}$ is non-zero. But, we will define it tobe stationary, the reason for which is given below.

## 4.2 Specific Force

The accelerometer measures the specific force f in the body frame b. This can be expressed as.

\\[ f^b = R^{bn}(a^n_{ii} − g^n ) \\]

where g denotes the gravity vector and $a^n_{ii}$ denotes the linear acceleration of the sensor expressed in the navigation frame, which is

\\[ a^n_{ii} = R^n_e R^e_i a^i_{ii} \\]

For navigation purposes, we are interested in the position of the sensor in the navigation frame $p^n$. After some manipulation, described in [Using Inertial Sensors for Position and Orientation Estimation](https://arxiv.org/pdf/1704.06053.pdf), we get,

\\[ a^n_{ii} = a^n_{nn} + 2\omega^i_e \times v^n_n + \omega^i_e \times \omega^i_e \times p^n \\]

where $a_{nn}$ is the acceleration required for navigation purposes. The term $\omega^i_e \times \omega^i_e \times p^n$ is known as the centrifugal acceleration and $2\omega^i_e \times v^n_n$ is known as the Coriolis acceleration. The centrifugal acceleration is typically absorbed in the (local) gravity vector. we discuss below the magnitude of both these terms.

The centrifugal acceleration depends on the location on the earth. It is possible to get a feeling for its magnitude by considering the property of the cross product stating that 

\\[ \vert\vert\omega^i_e \times \omega^i_e \times p^n\vert\vert_2 ≤ \vert\vert \omega^i_e \vert\vert_2 \vert\vert \omega^i_e \vert\vert_2 \vert\vert p^n \vert\vert_2 . \\]

Since the magnitude of $\omega^i_e$ ie is approximately $7.29 · 10^{−5}$ rad/s and the average radius of the earth is 6371 km, the magnitude of the centrifugal acceleration is less than or equal to $3.39 · 10^{−2} m/s^2$. 

The Coriolis acceleration depends on the speed of the sensor. Consider a person walking at a speed of 5 km/h. In that case the magnitude of the Coriolis acceleration is approximately $2.03 · 10^{−4} m/s^2$ .
For a car traveling at 120 km/h, the magnitude of the Coriolis acceleration is instead $4.86 · 10^{−3} m/s^2$.

# 5 Orientation Representation

### 5.1 Quaternion Representation

A commonly used parametrization of orientation is that of unit quaternions. A unit quaternion use a 4-dimensional representation of the orientation according to

\\[ q = (w \>\>\> x \>\>\> y \>\>\> z)^T = (w  \>\>\> v)^T,  \>\>\> q \in \mathbb{R}  \>\>\> and  \>\>\> \vert\vert q \vert\vert_2 = 1 \\]

where w is the scalar part and v is the vector component.

A unit quaternion is not a unique description of an orientation. The reason for this is that if q represents
a certain orientation, then −q describes the same orientation.

#### 5.1.1 Quaternion Addition

Addition is defined as the component-wise sum just like for a 4D vector. The sum is commutative (order is not important) and associative (grouping is not important).

\\[ q_1 + q_2 = w_1 + w_2 \>\>\> x_1 + x_2 \>\>\> y_1 + y_2 \>\>\> z_1 + z_2 \\]

#### 5.1.2 Quaternion multiplication

Quaternion multiplication is defined in multiple ways but the formula that I find the easiest to remember is:

\\[ q_1 \otimes q_2 = w_1w_2 - v_1.v_2 \>\>\> w_1v_2 + w_2v_1 + v_1 \times v_2 \\]

where $w_1$, $w_2$ are the scalar part of the quaternions $q_1, q_2$ and $v_1, v_2$ are the vector part of the quaternions $q_1, q_2$.

#### 5.1.3 Norm, Conjugate and Inverse

It’s also useful to define the norm of a quaternion as:

\\[ \vert\vert q \vert\vert = \sqrt{w^2 + x^2 + y^2 + z^2} \\]

a cojugate quaternion $q^∗$ that satisfies the property $q \otimes q^*$ = $\vert\vert q \vert\vert^2$ as

\\[ q^* = [w \>\>\> -x \>\>\> -y \>\>\> -z] \\]

and a quaternion inverse $q^{-1}$ with the property $q \otimes q^{-1} = [1 \>\> 0 \>\> 0 \>\> 0]$ as

\\[ q^{-1} = \frac{q^*}{\vert\vert q \vert\vert^2} \\]

#### Rotating a Vector

To rotate a 3D vctor $r$, we use the conjugate operation, as

\\[ \hat{r} \>\> = \>\> q \otimes [0 \>\> r] \otimes q^* \\]

#### Representing quaternion rotation as a matrix

Two quaternions $q_1 = [w_1 \>\> x_1 \>\> y_1 \>\> z_1]$ and $q_2 = [w_2 \>\> x_2 \>\> y_2 \>\> z_2]$, their multiplication can be represented as a matrix-vector product as 

\\[ q_1 \otimes q_2 = 
\begin{bmatrix}
w_2 & -x_2 & -y_2 & -z_2\\
x_2 & w_2 & z_2 & -y_2\\
y_2 & -z_2 & w_2 & x_2\\
z_2 & y_2 & -x_2 & w_2
\end{bmatrix}
\begin{bmatrix}
w_1 \\
x_1 \\
y_1 \\
z_1
\end{bmatrix}
\\]

# 6 Local Frame Navigation Equations

In this section, we will discuss the integration of gyroscope and accelerometer measurements to get orientation, position and velocity estimates.

#### 6.1 Orientation Estimation

As discussed above, the gyroscope measures the angular velocity $\omega_{ib}$ at each time instance t. However, its measurements are corrupted by a slowly time-varying bias $\delta_{\omega,t}$ and noise $e_{\omega,t}$. Hence, the gyroscope measurement model is given by

\\[ y_{\omega,t} = \omega^b_{ib} + \delta_{\omega,t} + e_{\omega,t} \\]

As was shown above, the gyroscope measurement noise is quite Gaussian. Because of this, it is typically assumed that $e^b_{\omega,t}$ ∼ $\mathcal{N(0, \Sigma_\omega)}$. If the sensor is properly calibrated, the measurements in the three gyroscope axes are independent. In that case, it can be assumed that

\\[ 
\begin{bmatrix}
\sigma^2_{\omega,x} & 0 & 0\\
0 & \sigma^2_{\omega,x} & 0\\
0 & 0 & \sigma^2_{\omega,x}
\end{bmatrix}
\\]


As described earlier, the angular velocity $\omega^b_{ib,t}$ can be expressed as 

\\[ {\omega^b}_{ib} = R^{bn}(\omega^n_{ie} + \omega^n_{en}) + \omega^b_{nb} \\]

If the sensor does not travel over significant distances as compared to the size of the earth — which
is the case for the application under discussion — the navigation frame n can safely be assumed to be stationary. In that case, the transport rate $\omega_{en,t}$ is zero. Although the earth rotation $\omega_{ie}$ as expressed in the body frame b is not constant, its magnitude as compared to the magnitude of the actual measurements is fairly small. Assuming that the earth rotation is negligible and the navigation frame is stationary leads to the following simplified measurement model,

\\[ y_{\omega,t} = \omega^b_{nb} + \delta_{\omega,t} + e_{\omega,t} \\]

Moreover, we know that the orientation and angular velocity $\omega^b_{nb}$ are related as 

\\[ \frac{dq^{nb} }{dt} = q^{nb} \otimes \frac{1}{2} \bar{\omega}^b_{nb}, \>\>\> \frac{dR^{nb}}{dt} = R^{nb} [\omega^b_{nb} \times] \\]

where $\bar{\omega}^b_{nb}$ means converting a 3d vector into a 4d vector by appending a 0 at the start, and $[\omega^b_{nb} \times]$ means a skew symmetric matrix of $\omega^b_{nb}$. 

Applying euler's discretization method for solving the above differential equation, we get

\\[ q^{nb}_{t+1} = q^{nb}_t \otimes exp_q(\frac{T}{2} \omega^b_{nb}), \>\>\>\> R^{nb}_{t+1} = R^{nb}_t exp_R( T\omega^b_{nb}) \\]

where, 
\\[
exp_q(\frac{T}{2} \omega^b_{nb}) = 
\begin{bmatrix}
cos(\frac{T}{2} \vert\vert \omega^b_{nb} \vert\vert)\\
\frac{\omega^b_{nb}}{\vert\vert \omega^b_{nb} \vert\vert} sin(\vert\vert \omega^b_{nb} \vert\vert)
\end{bmatrix}
\\]
and

\\[
exp_R( T\omega^b_{nb}) = I_3 + sin(\vert\vert T \omega^b_{nb} \vert\vert)
\begin{bmatrix}
\frac{\omega^b_{nb}}{\vert\vert \omega^b_{nb} \vert\vert} \times
\end{bmatrix}
+ (1 - cos(\vert\vert \omega^b_{nb} \vert\vert))
\begin{bmatrix}
\frac{\omega^b_{nb}}{\vert\vert \omega^b_{nb} \vert\vert} \times
\end{bmatrix}
^2
\\]

#### 6.2 Position and Velocity Estimation

The accelerometer measures the specific force $f^b_t$ at each time instance t. As shown above, the accelerometer measurements are typically assumed to be corrupted by a bias $\delta_{a,t}$ and noise $e_{a,t}$ as

\\[ y_{a,t} = f^b_t + \delta^b_{a,t} + e^b_{a,t} \\]

The accelerometer noise is typically quite Gaussian as was illustrated above and can hence be modeled as $e^b_{a,t}$ ∼ $\mathcal{N(0, \Sigma_a)}$. For a properly calibrated sensor, the covariance matrix $\Sigma_a$ can often be assumed to be diagonal. After some manipulation, our simplified accelerometer measurement model can be given as, 

\\[ y_{a,t} = R^{bn}(a^n_{nn} − g^n ) + \delta^b_{a,t} + e^b_{a,t} \\]

The fundamental continuous-time relations that form the basis of our dynamic models are the fact that
the position $p_n$ , velocity $v^n_n $ and acceleration $a^n_{nn}$ are related as,

\\[ v^n_n = \frac{dp^n}{dt}, \>\>\>\> a^n_{nn} = \frac{dv^n}{dt} \\]


Using euler's discretization method, we get,

\\[ p^n_{t+1} = p^n_{t} + Tv^n_{n,t} + \frac{T^2}{2}a^n_{nn,t}, \\
v^n_{n,t+1} = v^n_{n,t} + Ta^n_{nn,t}
\\]
